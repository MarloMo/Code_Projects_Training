## **Python Candlestick Chart**

Frances is asked to build a candlestick chart for her team to help everyone get an idea of how the daily pricing of AAPL changed over time, and what the price movements looked like during each day of trading.

A candlestick chart plots the daily opening price, closing price, lowest price and highest price of a particular stock, and shows how that price changed each day over a given period of time. Each day is shown by one “candlestick,” and can be helpful when analysts are trying to make predictions about how a particular price of a stock may move in the future.

```python
import datetime
import pandas as pnd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mplf
import mpl_finance as mpl_f
```

```python
# DataFrame (df)
df = pnd.read_csv("AAPL_data.csv")
print(df.head())

# Converts the Date column in df to datetime format using pnds
df["Date"] = pnd.to_datetime(df["Date"])
# Converts the datetime format to a format suitable for matplotlib candlestick
# chart (converts dates to numerical format expected by matplotlib)
df["Date"] = df["Date"].apply(mdates.date2num)

# Creates a new df containing the columns needed for the candlestick chart
candle_data = df[["Date", "Open", "High", "Low", "Close"]]
print(candle_data.head())

# Creates a new figure (f1) and axis (ax)
f1, ax = plt.subplots(figsize=(10, 5))
mpl_f.candlestick_ohlc(ax, candle_data.values, colorup="green", colordown="red")
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.title("Candlestick Chart for AAPL")
plt.xlabel("Date")
plt.ylabel("Value ($)")
```

Green and Red Candles: Green candles indicate price closing higher than opening (bullish day), while red candles indicate price closing lower than opening (bearish day).

Candlestick charts are valuable tools for technical analysis in finance, providing insights into price trends and market sentiment.
