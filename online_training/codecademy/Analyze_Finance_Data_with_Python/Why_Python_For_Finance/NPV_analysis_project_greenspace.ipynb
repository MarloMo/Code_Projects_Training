## **NPV Analysis - Project Greenspace**

During a weekly project meeting, one of your team members presents an exciting new project, Project Greenspace. After she presents, you have a chance to review the documents with the team and discuss some possible scenarios to help the team make a recommendation about whether the company should pursue the project.

You decide to perform a Net Present Value (NPV) analysis to determine the profitability of Project Greenspace. The Net Present Value determines the present value of the future cash inflows and outflows over the life the project.

```python
import matplotlib.pyplot as plt
```

```python
def calc_npv(rate, cash_flow):
    result = sum(cash_flow[t] / (1 + rate) ** t for t in range(len(cash_flow)))
    return result


project_greenspace_cashflow = [
    -2000000,
    0,
    0,
    250000,
    250000,
    500000,
    500000,
    750000,
    750000,
    1000000,
]

project_greenspace_cashflow_yr4_update = [
    -2000000,
    0,
    0,
    50000,
    250000,
    500000,
    500000,
    750000,
    750000,
    1000000,
]

project_greenspace_cashflow_15yr_update = [
    -2000000,
    0,
    0,
    50000,
    250000,
    500000,
    500000,
    750000,
    750000,
    500000,
    500000,
    500000,
    500000,
    500000,
    1000000,
]

discount_rate = [
    0.05,
    0.055,
    0.06,
    0.065,
    0.07,
    0.075,
    0.08,
    0.085,
    0.09,
    0.095,
    0.1,
    0.105,
    0.11,
    0.115,
    0.12,
    0.125,
    0.13,
    0.135,
    0.14,
    0.145,
    0.15,
]

net_present_value_greenspace_15yr_update = []
net_present_value_greenspace = []
net_present_value_greenspace_yr4_update = []
for rate in discount_rate:
    net_present_value_greenspace.append(calc_npv(rate, project_greenspace_cashflow))
    net_present_value_greenspace_yr4_update.append(
        calc_npv(rate, project_greenspace_cashflow_yr4_update)
    )
    net_present_value_greenspace_15yr_update.append(
        calc_npv(rate, project_greenspace_cashflow_15yr_update)
    )

plt.plot(
    discount_rate,
    net_present_value_greenspace_15yr_update,
    linewidth=3.0,
    color="purple",
    label="Project Greenspace : 15yr = $1mill",
)
plt.plot(
    discount_rate,
    net_present_value_greenspace_yr4_update,
    linewidth=3.0,
    color="red",
    label="Project Greenspace : yr4 = $50k",
)
plt.plot(
    discount_rate,
    net_present_value_greenspace,
    linewidth=3.0,
    color="green",
    label="Project Greenspace : yr4 = 250k",
)
plt.legend()
plt.axhline(y=0, linewidth=0.5, color="black")
plt.grid()
plt.title("NPV Profile: Project Greenspace")
plt.xlabel("Discount Rate")
plt.ylabel("New Presesnt Value")
```

 With a q5 cash flow of 1mil at year 15 the project starts to become unnatractive at a discount of 14%.
